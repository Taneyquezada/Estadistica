```python
#PROBLEMA 1

#1. Establezca una variable dependiente (Y) y una variable independiente (X)
import numpy as np

X = np.array([59, 92, 72, 90, 95, 87, 89, 77, 76, 65, 97, 42, 94, 62, 91]) #examenes cortos
Y = np.array([64, 84, 77, 80,77, 81, 80, 84, 80, 69, 83, 40, 78, 65, 90]) #promedio final

#2. Realice un diagrama de dispersión de datos
import matplotlib.pyplot as plt
plt.scatter(X, Y, color = "brown")
plt.xlabel("Exámenes cortos", color = "blue") #Nombrar eje x
plt.ylabel("Promedio final", color ="blue") #nombrar eje Y
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

#3. ¿Los datos soportan la suposición de linealidad?
print("Los datos si soportan la suposición de linealidad \n")

#4. calcular el coeficiente de correlación
Sxx = np.sum((X-np.mean(X))**2)
Sxy = np.sum((Y - np.mean(Y))*(X - np.mean(X)))
Syy = np.sum((Y - np.mean(Y))**2)

r = Sxy / np.sqrt(Sxx * Sxy)
print(f"Coeficiente de correlación: {r: 0.4f} \n")

#5. Calcule el coeficiente de determinación e interprete el resultado.
print(f"Coeficiente de determinación: {r**2: 0.4} \n")

#6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de
#dispersión.
b1 = Sxy / Sxx
b0 = np.mean(Y) - b1*np.mean(X)

fun = lambda x: b1*x + b0

Yc = fun(X)

plt.plot(X, Yc, color = "black", linestyle = "--")

#7. Obtenga un intervalo de confianza del 95% para la pendiene de la recta de
#regresión ajustada (b1)
nivel_de_significancia = 0.05
from scipy.stats import t
T = t.ppf(1 - nivel_de_significancia /2,df = X.size - 2)
se_b1 = np.sqrt(np.sum((Y - Yc)**2) / (Y.size - 2)) / np.sqrt(Sxx)
confianza_b1 = b1 - T * se_b1, b1 + T * se_b1
print(f"intervalo de confianza {1 - nivel_de_significancia: 0.0%}")
print(f"{confianza_b1[0] : 0.4f} < b1 < {confianza_b1[1] : 0.4f} \n")

#8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente,
#¿parece que se verifican los supuestos?
plt.figure()
residuales = Y - Yc
plt.scatter(X, residuales, color = "red")
plt.xlabel("Exámenes cortos", color = "blue")
plt.ylabel("Residuales", color = "blue")
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
plt.axhline(y = 0, color = "gray", linestyle = "--")
print ("No se cumplen los supuestos, ya que se puede notar una parábola en el gráfico de los residuales, por lo que se rompe el supuesto de linealidad \n")

#9. Realice la prueba de Shapiro para los residuales y comente el resultado.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f"valor -p para shapiro: {valor_p_sh: 0.4f} \n")

#11. Tres estudiantes sacaron 70, 75 y 84 de calificación. Según la recta de
#regresión ajustada, ¿Cuáles son los resultados esperados en estos tres alumnos?.
print(f"Para x = 70, y = {fun(70): 0.0f}")
print(f"Para x = 75, y = {fun(75): 0.0f}")
print(f"Para x = 84, y = {fun(84): 0.0f}")

#12. Realice una tabla ANOVA e interprete el resultado.
from statsmodels.formula.api import ols
# Y ~ X
modelo_2 = ols('Y ~ X', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_2)
tabla_anova
```

```python
#PROBLEMA 2
# 1. Establesca una variable dependiente ( Y ) y una variable independiente ( X ).
import numpy as np

X = np.array([11, 10, 8, 5, 9, 7, 3, 11, 8, 7, 2, 9, 8, 3]) #Salario
Y = np.array([18, 17, 29, 36, 11, 28, 35, 14, 20, 32, 39, 16, 31, 40]) #Ausencias

# 2. Realice un diagrama de dispersión para estos datos.
import matplotlib.pyplot as plt
plt.scatter(X, Y, color = "brown")
plt.xlabel("Salario", color = "blue") #Nombrar eje x
plt.ylabel("Ausencias", color ="blue") #nombrar eje Y
ax = plt.gca()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# 3. ¿Los datos soportan la suposición de linealidad?
print ("Si soportan la suposición de linealidad ya que se puede ver un poco como pasa una línea por los puntos. \n")

# 4. Calcule el coeficiente de correlación e interprete el resultado.
from scipy.stats import pearsonr
r,_= pearsonr(X, Y)
print(f'Coeficiente de correlación: {r:0.4f}')
print ("Por el valor de -0.8801 se puede determinar que su nivel de correlación es fuerte. \n")

# 5. Calcule el coeficiente de determinación e interprete el resultado.
print(f"El coeficiente  de determinación es: {r**2:0.4f}")
print("Quiere decir que existe un 77.46% que explica su variabilidad y la relación entre la variable independiente y la dependiente. \n")

#6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
import statsmodels.api as sm
x_constante = sm.add_constant(X)
modelo = sm.OLS(Y, x_constante).fit()
b0, b1 = modelo.params
fun = lambda x: b0 + b1 * x
Yc = fun(X)
plt.plot(X, Yc, color = 'black', linestyle = '--')
from sklearn.metrics import r2_score
r2 = r2_score(Y, Yc)

# 7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ( b1 )
import pandas as pd

nivel_de_confianza = 0.95
intervalo_de_confianza = pd.DataFrame(modelo.conf_int(alpha = 1 - nivel_de_confianza))
intervalo_de_confianza_b1 = intervalo_de_confianza.iloc[1]
print(f"Intervalo de confianza para b1 de {nivel_de_confianza: 0.0%}")
print(f"{intervalo_de_confianza_b1[0]:0.4f} < b1 < {intervalo_de_confianza_b1[1]: 0.4f}\n")

# 8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente, ¿Parece que se verifican los supuestos?
plt.figure()
residuales = Y - Yc
plt.scatter(X, residuales, color = 'purple')
plt.xlabel('Categoría de salario', color = "blue")
plt.ylabel('Ausencia', color = "blue")
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.axhline(y = 0, color = 'gray', linestyle = '--')
plt.show()
print("\n")

# 9. Realice la prueba de Shapiro para los residuales y comente el resultado.
from scipy.stats import shapiro
_, valor_p_sh = shapiro(residuales)
print(f"valor -p para shapiro: {valor_p_sh: 0.4f}")
print("Existe cierta desviación de los datos pero es una normalidad buena \n")

# 10. Realice la prueba de Brausch-Pagan para los residuales y comente el resultado.
from statsmodels.stats.api import het_breuschpagan
_, valor_p_bp, _, _ = het_breuschpagan( residuales, x_constante)
print(f"valor_p de Breush - pagan: {valor_p_bp: 0.4f}")
print("No existe heteroscedasticidad, por lo que cumple con su H0 \n")

# 11. Utiliza la recta de regresión para interpolar dos valores y extrapolar uno.
b0, b1 = modelo.params
x_interpolado1 = 6
y_interpolado1 = b0 + b1 *x_interpolado1
print(f'Para x = {x_interpolado1}, el valor interpolado de  y = {y_interpolado1: 0.4f}\n')

x_interpolado2 = 10
y_interpolado2 = b0 + b1 *x_interpolado2
print(f'Para x = {x_interpolado2}, el valor interpolado de  y = {y_interpolado2: 0.4f}\n')

x_extrapolado = 15
y_extrapolado = b0 + b1 *x_extrapolado
print(f'Para x = {x_extrapolado}, el valor extrapolado de  y = {y_extrapolado: 0.4f}\n')

# 12. Realice una tabla ANOVA e interprete el resultado.
from statsmodels.formula.api import ols
# Y ~ X
modelo_2 = ols('Y ~ X', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_2)
print("En esta tabla encontramos un valor -p menor 0.5, que es 0.000032,\n por lo que se relaciona este resultado con las ausencias. ")
tabla_anova

```
